```python
#Import dependencies to load data
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import sklearn
```

# Load Data

```python
#Load csv data
df = pd.read_csv("data.csv")
```

```python
#Preview data
df.head(5)
```

# Pre-Processing

```python
data = df.values
data2 = df.drop(["id","diagnosis","radius_se","texture_se","perimeter_se","area_se","smoothness_se","compactness_se","concavity_se",\
      "concave points_se","symmetry_se","fractal_dimension_se"],axis = 1)
```

```python
# Get training features
#X = data[:,2:12]
X = data2.values
```

```python
#Get label for training examples
Y = data[:,1]
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
label_encoder.fit(Y)
encoded_y = label_encoder.transform(Y)
```

```python
# Use train_test_split to create training and testing data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, encoded_y, random_state=1)
```

```python
# Create a StandardScater model and fit it to the training data
from sklearn.preprocessing import StandardScaler,MinMaxScaler
X_scaler = MinMaxScaler().fit(X_train)
```

```python
# Transform the training and testing data using the X_scaler
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
from tensorflow.keras.utils import to_categorical
# One-hot encoding
y_train_categorical = to_categorical(y_train)
y_test_categorical = to_categorical(y_test)
```

# Neural Network Model

## Build Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.regularizers import l2

model = Sequential()
model.add(Dense(units=40, activation='relu', input_dim=20,kernel_regularizer=l2(0.001), bias_regularizer=l2(0.001)))
model.add(Dense(units=40, activation='relu',kernel_regularizer=l2(0.001), bias_regularizer=l2(0.001)))
model.add(Dense(units=2, activation='softmax',kernel_regularizer=l2(0.001)))
```

## Compile Model

```python
# Compile the model
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

## Train Model

```python
# Fit the model to the training data
model.fit(X_train_scaled,y_train_categorical,epochs=1500,shuffle=True,verbose=2)
```

## Evaluate Model

```python
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```
